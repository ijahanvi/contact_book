```python
# Function to display the phone book records
def display_records():
    with open("phonebook.txt", "r") as file:
        data = file.readlines()
        for record in data:
            print(record.strip())

# Function to add a new contact record to the phone book
def add_record():
    name = input("Enter the contact name: ")
    number = input("Enter the contact number: ")
    with open("phonebook.txt", "a") as file:
        file.write(f"{name}: {number}\n")
    print("Contact added successfully.")

# Function to search for a contact record in the phone book
def search_record():
    name = input("Enter the contact name to search: ")
    found = False
    with open("phonebook.txt", "r") as file:
        data = file.readlines()
        for record in data:
            if name in record:
                print(record.strip())
                found = True
    if not found:
        print("Contact not found.")

# Function to edit a contact record in the phone book
def edit_record():
    name = input("Enter the contact name to edit: ")
    found = False
    with open("phonebook.txt", "r") as file:
        data = file.readlines()
    with open("phonebook.txt", "w") as file:
        for record in data:
            if name in record:
                number = input("Enter the new contact number: ")
                file.write(f"{name}: {number}\n")
                found = True
            else:
                file.write(record)
    if not found:
        print("Contact not found.")

# Function to delete a contact record from the phone book
def delete_record():
    name = input("Enter the contact name to delete: ")
    found = False
    with open("phonebook.txt", "r") as file:
        data = file.readlines()
    with open("phonebook.txt", "w") as file:
        for record in data:
            if name in record:
                found = True
            else:
                file.write(record)
    if found:
        print("Contact deleted successfully.")
    else:
        print("Contact not found.")

# Main program loop
while True:
    print("*** Phone Book Menu ***")
    print("Enter 1, 2, 3, 4, 5, or 6:")
    print("Enter 1 to display your contacts records")
    print("Enter 2 to add a new contact record")
    print("Enter 3 to search your contacts")
    print("Enter 4 to edit a contact record")
    print("Enter 5 to delete a contact record")
    print("Enter 6 to quit")
    choice = input("Choice: ")

    if choice == "1":
        display_records()
    elif choice == "2":
        add_record()
    elif choice == "3":
        search_record()
    elif choice == "4":
        edit_record()
    elif choice == "5":
        delete_record()
    elif choice == "6":
        break
    else:
        print("Invalid choice. Please try again.")
```
